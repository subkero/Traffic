```python
import requests
import pandas as pd

r=requests.get('https://covidtracking.com/api/v1/states/nc/daily.json')
df0=pd.read_json(r.text)
df0['date']=pd.to_datetime(df0['date'], format='%Y%m%d')
df0.set_index(['date'],inplace=True)
ax=df0.plot(kind='line', y='hospitalizedCurrently', color='DarkBlue', title='New Covid19 Deaths & Hopistilazations in NC by Date')
ax2=df0.plot(kind='line', y='deathIncrease', secondary_y=True,color='Red', ax=ax)
ax.set_ylabel('Currently Hospitalized')
ax2.set_ylabel('New Deaths')
ax.set_xlabel("Date")
```
